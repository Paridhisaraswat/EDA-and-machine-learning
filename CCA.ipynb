# CCA ------> complete case analysis



```python
# Complete case analysis is a missing value handling method where
#  rows containing missing values are analysis wether we can drop the missing value rows or we have to fill it .
```

```python
# df ----> missing data -----> filter missing data ------>
# new_df = filtered_columns_in_which_we_have_missing_value


# new_df = missing data frop
# Architecture  ----->histogram ------>
# past column in which we have missing data
# updated column in which we have no missing data


# if past data architecture is overlap to the new data architecture ---->
# it means we can drop missing data
# if past data architecture is not overlaping to the new data architecture ----->
# it means we can not drop missing data . we have to fill the missing data
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/dsjob.csv")
df.head()
```

```python
df.isnull().mean()*100
```

```python
# list comprehension
cols = [i for i in df.columns if df[i].isnull().mean() < 0.05 and df[i].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
df['education_level'].value_counts()
```

```python
# find the percentage of non missing data

len(df[cols].dropna())/len(df)
```

```python
new_df = df[cols].dropna()
df.shape , new_df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# original data
df['experience'].hist(bins = 50 , ax= ax, density = True , color = 'red')


# data after CCA  , the argument alpha makes the color transparent ,
# so we can see the overlay of the 2 distribution

new_df['experience'].hist(bins = 50 , ax = ax , color = 'green', density = True , alpha = 0.7)
```
