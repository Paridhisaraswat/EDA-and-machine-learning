```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4 , n_features= 1,
                      n_informative= 1,
                      n_targets= 1,
                      noise = 80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
LinearRegression(copy_X = True ,
                 fit_intercept= True ,
                 n_jobs = None)
```

```python
reg.coef_     # m slope
```

```python
reg.intercept_    # b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red')
```

```python
# let's apply Gradient Descent assuming Slope is
# Constant = 78.35
# and let's assume the strating value for
# intercept b = 0

y_pred = ((78.35 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',
         label = 'OLS')

plt.plot(x,y_pred , color = 'green',label = 'b = 0')
plt.legend()
plt.show()
```

```python
import numpy as np

arr = np.array([[1,2,3],[4,5,6]])

# using ravel() function to flatten the array

flattened_arr = arr.ravel()

print(arr)
print(flattened_arr)
```

ravel() is a function in Numpy that returns a falttened array i.e. it return a contiguous flattened array row - major order

in other words it return a !D array by flattening the i/p array ,whether it is a multi-dimensional or one dimensional array

```python
m = 78.35
b = 0
loss_slope = -2*np.sum(y - m*x.ravel()-b)         ### dl/db
loss_slope
```

```python
# let's take learning rate = 0.1
lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
# calculating the new intercept

b = b-step_size        ### b_new = b_old - lr * dl/db
b
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)


plt.scatter(x,y)
plt.plot(x, reg.predict(x),color = 'red',label = 'OLS')
plt.plot (x,y_pred1,color = 'black', label = 'b = {}'.format(b))
plt.plot(x,y_pred,color = 'yellow',label = 'b = 0')

plt.legend()
plt.show()
```

```python
# Iteration 2

loss_slope = -2* np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b -step_size
b
```

```python
y_pred2 = ((78.35 * x) + b).reshape(4)


plt.scatter(x,y)
plt.plot(x, reg.predict(x),color = 'red',label = 'OLS')
plt.plot (x,y_pred2,color = 'green', label = 'b = {}'.format(b))
plt.plot (x,y_pred1,color = 'blue', label = 'b = {}'.format(b))
plt.plot(x,y_pred,color = 'yellow',label = 'b = 0')

plt.legend()
plt.show()
```

```python
b = b -step_size
b
```
