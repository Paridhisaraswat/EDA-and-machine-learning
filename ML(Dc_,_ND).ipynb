```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop(columns =["city","cough"])
```

```python
df.head()
```

```python
df["has_covid"] = df['has_covid'].map({"Yes":1,"No":0})
df['gender'] = df["gender"].map({"Female":1,"Male":0})
```

```python
df.head()
```

```python
x = df.drop(columns =["has_covid"])
y = df["has_covid"]
```

```python
# that splits your dataset into training and testing subsets
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
np.round(x_train.describe() , 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
# fit means learn from the parameters and transfrom means apply on that data
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x_train.columns)
```

```python
np.round(x_train_new.describe() , 2)
```
