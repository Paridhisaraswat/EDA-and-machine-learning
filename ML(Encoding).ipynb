#ENCODING

```python
# ENCODING ---> transform your categorical data into numerical data
# (1) Label ----> it will encode data 1-1 columns .
# (2) Ordinal ---> it will encode data more than 1 columns .
# (3) Onehot -----> it will create new column based on sub-categories but it has only  value not column name
# (4) get_dummies ----> it will create new column based on sub categories but it show column name with values
```

#oneHot encoding
column has_covid ['y' ,'n'] ----> new column y_has_covid , n_has_covid ----> but contain only values not column name


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/covid_toy.csv")
df.head()
```

```python
df = df.dropna()
```

```python
df = df.drop(columns =['age','fever'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first',sparse_output = False , dtype = np.int32)
#sparse_output =>  to return a dense NumPy array
```

```python
# gender ---> 2 -1 = 1
# cough ----> 2- 1 = 1
# city -----> 4- 1 = 3
# has_covid ---> 2- 1 = 1
```

```python
ohe.fit_transform(df[['gender','cough','city','has_covid']])
```

```python
a = ohe.fit_transform(df[['gender','cough','city','has_covid']])
```

```python
# Convert to DataFrame
encoded_df = pd.DataFrame(a,
    columns=ohe.get_feature_names_out(['gender','cough','city','has_covid']))

encoded_df.head()
```

#get_dummies -----> pandas method

```python
a = pd.get_dummies(df , columns = ['gender','cough','city','has_covid'], drop_first=True)
```

```python
a = a.astype(int)
a
```
